```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

### 1. In how many cities Summer Olympics is held so far?



```python
df=pd.read_csv('summer.csv')
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)


```python
data=[]
gold=df[df['Medal']=='Gold']
for sport in gold['Sport'].unique():
  data.append([sport,len(gold[gold['Sport']== sport])])
pd.DataFrame(data,columns=['Sport','No. of Gold']).sort_values(by='No. of Gold', ascending=False).head().plot(x='Sport',y='No. of Gold',kind='bar',figsize=(10,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)


```python
df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot(x='Sports',y='No. of Medals',kind='bar',figsize=(10,5))
```

### 4. Which player has won most number of medals? (Top 5)


```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot.bar()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)


```python
data=[]
gold=df[df['Medal']=='Gold']
for Athlete in gold['Athlete'].unique():
  data.append([Athlete,len(gold[gold['Athlete']== Athlete])])
pd.DataFrame(data,columns=['Athlete','No. of Gold']).sort_values(by='No. of Gold', ascending=False).head().plot(x='Athlete',y='No. of Gold',kind='bar',figsize=(10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?


```python
india=df[df['Country']=='IND']
india_gold=india[india['Medal']=='Gold']
data=[]
for year in india_gold['Year'].unique():
  data.append(year)
data[0]
```

### 7. Which event is most popular in terms on number of players? (Top 5)


```python
data=[]
for event in df['Event'].unique():
  data.append([event,len(df[df['Event']==event])])
pd.DataFrame(data,columns=['Event','No. of Athlete']).sort_values(by='No. of Athlete', ascending=False).head().plot(x='Event',y='No. of Athlete',kind='bar',figsize=(10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)


```python
female=df[df['Gender']=='Women']
female_gold=female[female['Medal']=='Gold']
data=[]
for sport in female_gold['Sport'].unique():
  data.append([sport,len(female_gold[female_gold['Sport']==sport])])
pd.DataFrame(data,columns=['Sport','No.ofFemaleGoldmedlist']).sort_values(by='No.ofFemaleGoldmedlist', ascending=False).head().plot(x='Sport',y='No.ofFemaleGoldmedlist',kind='bar',figsize=(10,5))
```
